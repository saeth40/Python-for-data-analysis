```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Matplotlib


```
x=np.arange(0,20)
y=np.arange(len(x))
plt.scatter(x,y,c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
#plt.savefig('Test.png')
```

```
y=2*(x^2) + 5
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y=2x^2+5')
```

```
#subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```
x = np.arange(20) 
y = 5 * x + 5 
plt.title("y = 5x + 5") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.plot(x,y) 
plt.show()
```

```
x = np.arange(0, 5 * np.pi, 0.1) 
x
```

```
y = np.sin(x) 
plt.title("sine") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show()
```

```
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'b--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'r--') 
plt.title('Cos')  
   
# Show the figure. 
plt.show()
```

```
#bar plot
x = [1,3,5] 
y = [11,16,9]  

x2 = [4,6,8] 
y2 = [6,15,7] 
plt.bar(x, y,color = 'r') 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```
#histograms
a = np.random.rand(100) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```
#box plot
data = [np.random.normal(0, std, 20) for std in range(1, 5)]
data
```

```
plt.boxplot(data,vert=True,patch_artist=False);
```

```
#pie chart

# Data to plot
labels = 'Type1', 'Type2', 'Type3', 'Type4'
sizes = [100, 130, 125, 145]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.3, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```

# Seaborn

```
import seaborn as sns
import pandas as pd
from pandas import Series, DataFrame
```

```
#load data (fifa19)
df=pd.read_csv('fifa19.csv')
df.head()
```

```
df.columns
```

```
#line plot
#50 rows
df2=df.iloc[:50]
plt.figure(figsize=(16,6))
sns.lineplot(data=df2)
```

```
plt.figure(figsize=(14,15))
plt.title("Overall vs Dollars")
plt.xlabel("Overall")
plt.ylabel("Dollars")
sns.lineplot(data=df2, x='Overall',y='Wage', label="Wage")
sns.lineplot(data=df2, x='Overall',y='Value', label="Value")
```

```
#Bar Charts and Heatmaps
# Set the width and height of the figure
plt.figure(figsize=(15,6))

# Add title
plt.title("index vs special")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=df2['Unnamed: 0'], y=df2['Special'])

# Add label for vertical axis
plt.xlabel("index")
plt.ylabel("Special")
```

```
plt.title("title")
df3=df.iloc[:10]
# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=df3[['ID','Special','Unnamed: 0','Age','Overall']], annot=True)

# Add label for horizontal axis
plt.xlabel("X")
plt.ylabel("Y")
```

```
#Scatter Plots
plt.figure(figsize=(14,10))
sns.scatterplot(x=df2['Unnamed: 0'],y=df2['Special'])
```

```
#find trend line
plt.figure(figsize=(14,10))
sns.regplot(x=df2['Unnamed: 0'],y=df2['Special'])
```

```
#hue = catagorized by
plt.figure(figsize=(14,10))
sns.scatterplot(x=df2['Unnamed: 0'], y=df2['Special'], hue=df2['Preferred Foot'])
```

```
#find mor than 1 trand line
sns.lmplot(x="Unnamed: 0", y="Special", hue="Preferred Foot", data=df2)
```

```
#categorical scatter plot
sns.swarmplot(x=df2['Preferred Foot'],
              y=df2['Special'])
```

```
#Distributions
#histogram
sns.distplot(a=df2['Overall'], kde=False)
```

```
#Density plots
sns.kdeplot(df2['Overall'], shade=True)
```

```
#2D KDE plots
sns.jointplot(x=df2['Overall'], y=df2['Special'], kind="kde")
```

```
sns.jointplot(x=df2['Overall'], y=df2['Special'], kind="hex")
```

```
sns.jointplot(x=df2['Overall'], y=df2['Special'], kind="reg")
```

```
# Histograms for each 
sns.distplot(a=df['Overall'], label="Overall", kde=False)
sns.distplot(a=df['Potential'], label="Potential", kde=False)
sns.distplot(a=df['Age'], label="Age", kde=False)

# Add title
plt.title("Histogram of 3 catagories")

# Force legend to appear
plt.legend()
```

```
sns.kdeplot(data=df['Overall'], label="Overall", shade=True)
sns.kdeplot(data=df['Potential'], label="Potential", shade=True)
sns.kdeplot(data=df['Age'], label="Age", shade=True)

# Add title
plt.title("Histogram of 3 catagories")
plt.legend()
```
